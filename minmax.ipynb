minmax 1

```python
# Minmax1: Find minimum and maximum in one pass
n = int(input("Enter N: "))
if n <= 0:
    print("Invalid input")
else:
    current = float(input("Enter number 1: "))
    min_val = current
    max_val = current

    for i in range(2, n + 1):
        num = float(input(f"Enter number {i}: "))
        if num < min_val:
            min_val = num
        if num > max_val:
            max_val = num

    print(f"Minimum: {min_val}")
    print(f"Maximum: {max_val}")
```

minmax 2

```python
# Minmax2: Find minimal rectangle area
n = int(input("Enter N: "))
if n <= 0:
    print("Invalid input")
else:
    a, b = map(float, input("Enter first rectangle sides (a b): ").split())
    min_area = a * b

    for i in range(2, n + 1):
        a, b = map(float, input(f"Enter rectangle {i} sides (a b): ").split())
        area = a * b
        if area < min_area:
            min_area = area

    print(f"Minimal area: {min_area}")
```

minmax 3

```python
# Minmax3: Find maximal perimeter
n = int(input("Enter N: "))
if n <= 0:
    print("Invalid input")
else:
    a, b = map(float, input("Enter first rectangle sides (a b): ").split())
    max_perimeter = 2 * (a + b)

    for i in range(2, n + 1):
        a, b = map(float, input(f"Enter rectangle {i} sides (a b): ").split())
        perimeter = 2 * (a + b)
        if perimeter > max_perimeter:
            max_perimeter = perimeter

    print(f"Maximal perimeter: {max_perimeter}")
```

minmax 4

```python
# Minmax4: Find index of minimal element
n = int(input("Enter N: "))
if n <= 0:
    print("Invalid input")
else:
    min_val = float(input("Enter number 1: "))
    min_index = 1

    for i in range(2, n + 1):
        num = float(input(f"Enter number {i}: "))
        if num < min_val:
            min_val = num
            min_index = i

    print(f"Index of minimum: {min_index}")
```

minmax 5

```python
# Minmax5: Find part with maximal density (m/v)
n = int(input("Enter N: "))
if n <= 0:
    print("Invalid input")
else:
    m, v = map(float, input("Enter mass and volume (m v) for part 1: ").split())
    max_density = m / v
    max_index = 1

    for i in range(2, n + 1):
        m, v = map(float, input(f"Enter mass and volume (m v) for part {i}: ").split())
        density = m / v
        if density > max_density:
            max_density = density
            max_index = i

    print(f"Part {max_index} has maximal density: {max_density:.2f}")
```
