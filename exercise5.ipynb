```python
# Problem 1: Sum and product of first 5 elements
arr = list(map(float, input("Enter array elements (space-separated): ").split()))
first_five = arr[:5]  # Takes up to first 5 elements

total_sum = sum(first_five)
product = 1
for num in first_five:
    product *= num

print(f"Sum: {total_sum}")
print(f"Product: {product}")
```

```python
# Problem 2: Create array of 2^1 to 2^N
N = int(input("Enter N: "))
powers = [2 ** (i+1) for i in range(N)]  # 2^1, 2^2, ..., 2^N
print(f"Powers of two: {powers}")
```

```python
# Problem 3: Create cumulative sum array B
A = list(map(float, input("Enter array A (space-separated): ").split()))
B = []
current_sum = 0

for k in range(len(A)):
    current_sum += A[k]
    B.append(current_sum)

print(f"Cumulative sums: {B}")
```

```python
# Problem 4: Sum and product of even elements in 2D array
n = int(input("Enter rows (n): "))
m = int(input("Enter cols (m): "))
array = []
even_sum = 0
even_product = 1
has_even = False

for i in range(n):
    row = list(map(int, input(f"Enter row {i+1} ({m} numbers): ").split()))
    array.append(row)
    for num in row:
        if num % 2 == 0:
            even_sum += num
            even_product *= num
            has_even = True

if not has_even:
    even_product = 0  # No even numbers case

print(f"Sum of evens: {even_sum}")
print(f"Product of evens: {even_product}")
```
