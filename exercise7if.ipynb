```python
# Problem 19: Swap A and B to ensure A <= B
A = float(input("Enter A: "))
B = float(input("Enter B: "))

if A > B:
    A, B = B, A  # Swap values

print(f"New A (smaller): {A}")
print(f"New B (larger): {B}")
```

```python
# Problem 10: Update A and B based on equality
A = int(input("Enter A: "))
B = int(input("Enter B: "))

if A != B:
    new_value = A + B
    A = new_value
    B = new_value
else:
    A = 0
    B = 0

print(f"New A: {A}")
print(f"New B: {B}")
```

```python
# Problem 11: Assign max value or zero
A = int(input("Enter A: "))
B = int(input("Enter B: "))

if A != B:
    max_val = max(A, B)
    A = max_val
    B = max_val
else:
    A = 0
    B = 0

print(f"New A: {A}")
print(f"New B: {B}")
```
