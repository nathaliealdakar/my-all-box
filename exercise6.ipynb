```python
# Problem 1: Count a character in each line of a poem
poem = input("Enter the poem (press Enter after each line, leave empty to finish):\n")
lines = poem.split('\n')  # Split input into lines
char = input("Enter the character to count: ").strip()

# Ensure input is a single character
if len(char) != 1:
    print("Error: Enter exactly one character")
else:
    for i, line in enumerate(lines, 1):
        count = line.count(char)
        print(f"Line {i}: {count} occurrences")

```

```python
# Problem 2: Count words starting with 'b'
text = input("Enter English text: ").split()
count = 0

for word in text:
    if word.startswith('b'):  # Case-sensitive check
        count += 1

print(f"Number of words starting with 'b': {count}")
```

```python
# Problem 3: Decrypt reversed words
encrypted = input("Enter encrypted message: ")
words = encrypted.split()
decrypted = [word[::-1] for word in words]  # Reverse each word
result = ' '.join(decrypted)

print(f"Decrypted message: {result}")
```
