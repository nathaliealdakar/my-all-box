```python
# Problem 1: Compare which number has more digits
def count_digits(number):
    return len(str(abs(number)))  # Ignore negative sign

num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))

digits1 = count_digits(num1)
digits2 = count_digits(num2)

if digits1 > digits2:
    print("First number has more digits")
elif digits2 > digits1:
    print("Second number has more digits")
else:
    print("Both numbers have the same number of digits")
```

```python
# Problem 2: Find GCD and LCM of two numbers
def gcd(a, b):
    while b != 0:
        a, b = b, a % b
    return a

a = int(input("Enter first number: "))
b = int(input("Enter second number: "))

gcd_value = gcd(a, b)
lcm_value = (a * b) // gcd_value  # Integer division

print(f"GCD: {gcd_value}")
print(f"LCM: {lcm_value}")
```

```python
# Problem 3: Count digits using recursion
def count_digits_recursive(n):
    if n < 10:
        return 1
    else:
        return 1 + count_digits_recursive(n // 10)

number = int(input("Enter a natural number: "))
if number <= 0:
    print("Invalid input (must be > 0)")
else:
    digit_count = count_digits_recursive(number)
    print(f"Number of digits: {digit_count}")
```
