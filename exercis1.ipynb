```
# Problem 1: Simplify (5x -10y)/(2y) ÷ (6y -3x)/(8y²)
x = float(input("Enter x: "))
y = float(input("Enter y: "))

# Calculate numerator and denominator of the first fraction
numerator1 = 5 * x - 10 * y
denominator1 = 2 * y

# Calculate numerator and denominator of the second fraction
numerator2 = 6 * y - 3 * x
denominator2 = 8 * y ** 2

# Division becomes multiplication by reciprocal
result = (numerator1 / denominator1) * (denominator2 / numerator2)

print(f"Result: {round(result, 2)}")
```

```
# Problem 2: Calculate trigonometric expression
import math

a = float(input("Enter angle a (in radians): "))
b = float(input("Enter angle b (in radians): "))

# Calculate numerator and denominator
numerator = math.sin(a + b) - math.cos(a) * math.sin(b)
denominator = math.cos(a - b) - math.sin(a) * math.sin(b)

if denominator == 0:
    print("Error: Division by zero")
else:
    c = numerator / denominator
    print(f"Result: {round(c, 2)}")
```

```
import math

# Input the side length of the equilateral triangle
side = float(input("Enter the side length of the equilateral triangle: "))

# Calculate the area
area = (math.sqrt(3) / 4) * (side ** 2)

# Calculate the height
height = (math.sqrt(3) / 2) * side

# Calculate the inradius (radius of inscribed circle)
inradius = (math.sqrt(3) / 6) * side

# Calculate the circumradius (radius of circumscribed circle)
circumradius = (math.sqrt(3) / 3) * side

# Print results rounded to 2 decimal places
print(f"\nResults for a triangle with side {side}:")
print(f"Area: {round(area, 2)}")
print(f"Height: {round(height, 2)}")
print(f"Inradius: {round(inradius, 2)}")
print(f"Circumradius: {round(circumradius, 2)}")
```

```
# Problem 5: Check if Queen can move in one step
x1 = int(input("Enter x1 (1-8): "))
y1 = int(input("Enter y1 (1-8): "))
x2 = int(input("Enter x2 (1-8): "))
y2 = int(input("Enter y2 (1-8): "))

# Check same row, column, or diagonal
same_row = (x1 == x2)
same_col = (y1 == y2)
same_diagonal = (abs(x1 - x2) == abs(y1 - y2))

can_move = same_row or same_col or same_diagonal
print(f"Queen can move: {can_move}")
```
