```python
# Problem 1: Calculate total distance over 7 days
daily_distance = 10  # First day distance
total = daily_distance

for day in range(2, 8):  # Days 2 to 7
    daily_distance *= 1.1  # Increase by 10%
    total += daily_distance

print(f"Total distance after 7 days: {round(total, 2)} km")
```

```python
# Problem 2: Calculate S = 1 + sin(1) + sin(2) + ... + sin(N)
import math

N = int(input("Enter N: "))
sum_sines = 1.0  # Start with the initial [1]

for k in range(1, N + 1):
    sum_sines += math.sin(k)

print(f"Result: {round(sum_sines, 4)}")
```

```python
# Problem 3: Sum series until term < 0.001
epsilon = 0.001
total = 0.0
n = 1

while True:
    term = 1 / (1 + 3 * n)
    if term < epsilon:
        break
    total += term
    n += 1

print(f"Sum: {round(total, 3)}")
```

```python
# Problem 4: Calculate sum and product of N numbers
N = int(input("Enter N: "))
numbers = []

for _ in range(N):
    num = float(input("Enter a number: "))
    numbers.append(num)

total_sum = sum(numbers)
product = 1.0
for num in numbers:
    product *= num

print(f"Sum: {total_sum}")
print(f"Product: {product}")
```
