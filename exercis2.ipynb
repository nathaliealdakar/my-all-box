```python
# Problem 1: Calculate f(x) based on conditions
x = float(input("Enter x: "))

if x >= 3:
    result = 2 * x + 9  # Simplified from -x + 3x + 9
else:
    denominator = x**3 - 1
    if denominator == 0:
        print("Error: Division by zero")
        result = None
    else:
        result = x / denominator

print(f"f(x) = {round(result, 2) if result is not None else 'undefined'}")
```

```python
# Problem 2: Determine which point is closer to the origin
x1 = float(input("Enter x1: "))
y1 = float(input("Enter y1: "))
x2 = float(input("Enter x2: "))
y2 = float(input("Enter y2: "))

distance_sq_a = x1**2 + y1**2  # Avoid sqrt for simplicity
distance_sq_b = x2**2 + y2**2

if distance_sq_a < distance_sq_b:
    print("Point A is closer to the origin")
elif distance_sq_a > distance_sq_b:
    print("Point B is closer to the origin")
else:
    print("Both points are equally distant")
```

```python
# Problem 3: Check equality to d or compute max(d - values)
m = float(input("Enter m: "))
n = float(input("Enter n: "))
p = float(input("Enter p: "))
d = float(input("Enter d: "))

equal_to_d = []
if m == d:
    equal_to_d.append("m")
if n == d:
    equal_to_d.append("n")
if p == d:
    equal_to_d.append("p")

if equal_to_d:
    print(f"Variables equal to d: {', '.join(equal_to_d)}")
else:
    max_diff = max(d - m, d - n, d - p)
    print(f"Maximum difference: {round(max_diff, 2)}")
```

```python
# Problem 4: Find max and min of two distinct numbers
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))

print(f"Maximum: {max(num1, num2)}")
print(f"Minimum: {min(num1, num2)}")
```
